## This is my Psuedo Code 
Pseudo code notes: 
1. Start
2. Create an empty shopping cart list
3. Initialize total_cost to 0
4. Loop:
   a. Display menu options to the user:
      - 1. Add item to the cart
      - 2. Remove item from the cart
      - 3. View cart
      - 4. Checkout
      - 5. Exit
   b. Read the user's choice
   c. If choice is 1:
      i. Read the item name and price from the user
      ii. Add the item to the shopping cart
      iii. Update total_cost
   d. If choice is 2:
      i. Display the current cart and ask the user which item to remove
      ii. Remove the selected item from the cart
      iii. Update total_cost
   e. If choice is 3:
      i. Display the current cart and the total cost
   f. If choice is 4:
      i. Display the final cart and total_cost
      ii. End the loop
   g. If choice is 5:
      i. End the loop
5. End


```python
def display_menu():
    print("Menu Options:")
    print("1. Add item to the cart")
    print("2. Remove item from the cart")
    print("3. View cart")
    print("4. Checkout")
    print("5. Exit")

def add_item(cart, total_cost):
    item_name = input("Enter item name: ")
    item_price = float(input("Enter item price: "))
    cart.append((item_name, item_price))
    total_cost += item_price
    return cart, total_cost

def remove_item(cart, total_cost):
    print("Current Cart:")
    for i, (name, price) in enumerate(cart):
        print(f"{i + 1}. {name} - ${price}")
    
    item_to_remove = int(input("Enter the number of the item to remove: ")) - 1
    if 0 <= item_to_remove < len(cart):
        total_cost -= cart[item_to_remove][1]
        del cart[item_to_remove]
    
    return cart, total_cost

def view_cart(cart, total_cost):
    print("Current Cart:")
    for i, (name, price) in enumerate(cart):
        print(f"{i + 1}. {name} - ${price}")
    print(f"Total Cost: ${total_cost}")

def checkout(cart, total_cost):
    print("Final Cart:")
    for i, (name, price) in enumerate(cart):
        print(f"{i + 1}. {name} - ${price}")
    print(f"Total Cost: ${total_cost}")

def main():
    shopping_cart = []
    total_cost = 0

    while True:
        display_menu()
        choice = int(input("Enter your choice: "))

        if choice == 1:
            shopping_cart, total_cost = add_item(shopping_cart, total_cost)
        elif choice == 2:
            shopping_cart, total_cost = remove_item(shopping_cart, total_cost)
        elif choice == 3:
            view_cart(shopping_cart, total_cost)
        elif choice == 4:
            checkout(shopping_cart, total_cost)
            break
        elif choice == 5:
            break

if __name__ == "__main__":
    main()

```
